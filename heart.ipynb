```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

# Null Values

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
mean = ['Blood Pressure', 'Cholesterol Level', 'BMI', 'CRP Level', 'Homocysteine Level']

for i in data.columns :
      for j in mean:
            if i == j:
                  data[i].fillna(value=np.mean(data[i]), inplace=True)
            else: 
                data[i].fillna(value=data[i].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
data.shape
```

# Encoding 

```python
data.columns
```

```python
print(data['Exercise Habits'].unique())
print(data['Alcohol Consumption'].unique())
print(data['Stress Level'].unique())
print(data['Sugar Consumption'].unique())
```

```python
data.info()
```

```python
gender = pd.get_dummies(data=data['Gender'], dtype=int, drop_first=True, prefix='Gender')
excercise = pd.get_dummies(data=data['Exercise Habits'], dtype=int, drop_first=True, prefix='Excercise')
smoking = pd.get_dummies(data=data['Smoking'], dtype=int, drop_first=True, prefix='Smoking')
family = pd.get_dummies(data=data['Family Heart Disease'], dtype=int, drop_first=True, prefix='Family')
diabetes = pd.get_dummies(data=data['Diabetes'], dtype=int, drop_first=True, prefix='Diabetes')
data.drop(columns=['High Blood Pressure'], inplace=True)
hdl = pd.get_dummies(data=data['Low HDL Cholesterol'], dtype=int, drop_first=True, prefix='HDL')
ldl = pd.get_dummies(data=data['High LDL Cholesterol'], dtype=int, drop_first=True, prefix='LDL')
alcohol = pd.get_dummies(data=data['Alcohol Consumption'], dtype=int, drop_first=True, prefix='Alcohol')
stress = pd.get_dummies(data=data['Stress Level'], dtype=int, drop_first=True, prefix='Stress')
sugar = pd.get_dummies(data=data['Sugar Consumption'], dtype=int, drop_first=True, prefix='Sugar')
heart = pd.get_dummies(data=data['Heart Disease Status'], dtype=int, drop_first=True, prefix='Heart')
```

```python
data = pd.concat([data, gender, excercise, smoking, family, diabetes, hdl, ldl, alcohol, stress, sugar, heart], axis=1)
data.drop(columns=['Gender', 'Exercise Habits', 'Smoking', 'Family Heart Disease', 'Diabetes', 'Low HDL Cholesterol', 'High LDL Cholesterol', 'Alcohol Consumption', 'Stress Level', 'Sugar Consumption', 'Heart Disease Status'], inplace=True)
```

## Threshold Tuning(Gradient Boost)

```python
y_prob = gbc.predict_proba(x_test)[:, 1]
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

y_scores = gbc.predict_proba(x_test)[:,1]
precision, recall, thresholds = precision_recall_curve(y_test, y_scores)

plt.plot(thresholds, precision[:-1], label='Precision')
plt.plot(thresholds, recall[:-1], label='Recall')
plt.xlabel('Threshold')
plt.legend()
plt.title('Precision vs Recall vs Threshold')
plt.show()
```

```python
y_pred = (y_prob > 0.58).astype(int)
```

```python
print(classification_report(y_test, y_pred))
```
